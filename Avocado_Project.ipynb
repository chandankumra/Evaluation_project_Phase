```python
x1=scale.drop(columns='region')
y1=y
```

```python
y1
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif
```

```python
x1.drop(columns='Small_Bags',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=.25,random_state=5)
```

```python
print(x_train.shape,'____',x_test.shape,'_____',y_train.shape,'______',y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
rfc.score(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

```python
accuracy_score(y_pred,y_test)
```
