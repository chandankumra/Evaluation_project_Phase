# whether or not it will rain tomorrow

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
x=data.drop(columns='RainTomorrow')

y=data.RainTomorrow
```

```python
scaled=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['column']=scaled.columns
vif['values']=[variance_inflation_factor(scaled.values,i) for i in range(len(scaled.columns))]
vif
```

```python
scaled.drop(columns='Temp3pm',inplace=True)
```

```python
vif=pd.DataFrame()
vif['column']=scaled.columns
vif['values']=[variance_inflation_factor(scaled.values,i) for i in range(len(scaled.columns))]
vif
```

```python
scaled.drop(columns='Temp9am',inplace=True)
```

```python
vif=pd.DataFrame()
vif['column']=scaled.columns
vif['values']=[variance_inflation_factor(scaled.values,i) for i in range(len(scaled.columns))]
vif
```

```python
scaled.drop(columns='Pressure9am',inplace=True)
```

```python
vif=pd.DataFrame()
vif['column']=scaled.columns
vif['values']=[variance_inflation_factor(scaled.values,i) for i in range(len(scaled.columns))]
vif
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled,y,test_size=.25,random_state=35)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred=lr.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor(n_estimators=1000,max_depth=5,random_state=12)
```

```python
gbr.fit(x_train,y_train)
```

```python
gbr.score(x_train,y_train)
```

```python
gbr.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
error=[mean_squared_error(y_test,y_pred) for y_pred in gbr.staged_predict(x_test)]
```

```python
error
```

```python
bst_esm=np.argmin(error)
```

```python
bst_esm
```

```python
gb_b=GradientBoostingRegressor(n_estimators=bst_esm,max_depth=5)
```

```python
gb_b.fit(x_train,y_train)
```

```python
gb_b.score(x_train,y_train)
```

```python
gb_b.score(x_test,y_test)
```
