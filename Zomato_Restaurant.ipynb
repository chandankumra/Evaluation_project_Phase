```python
Votes_h=q3.Votes+(1.5*iqr.Votes)
ind_num=np.where(Votes_h<data['Votes'])
data=data.drop(data.index[ind_num])
data.reset_index()
data.shape
```

```python
ag_l=q3.Aggregate_rating-(1.5*iqr.Aggregate_rating)
ind_num=np.where(ag_l>data['Aggregate_rating'])
data=data.drop(data.index[ind_num])
data.reset_index()
data.shape
```

# Price_range

```python
data_pr=data
```

```python
acft_h=q3.Average_Cost_for_two+(1.5*iqr.Average_Cost_for_two)
ind_num=np.where(acft_h<data['Average_Cost_for_two'])
data_pr=data_pr.drop(data_pr.index[ind_num])
data_pr.reset_index()
data_pr.shape
```

```python
x=data_pr.drop('Price_range',axis=1)
y=data_pr.Price_range
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
scaled=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
scaled.head()
```

```python
vif=pd.DataFrame()
vif['values']=[variance_inflation_factor(scaled,i) for i in range(len(scaled.columns))]
vif['feature']=scaled.columns
vif
```

```python
scaled.drop(columns=['Country_Code','Country','Currency','Switch_to_order_menu','Locality_Verbose'],inplace=True)
```

```python
vif=pd.DataFrame()
vif['values']=[variance_inflation_factor(scaled,i) for i in range(len(scaled.columns))]
vif['feature']=scaled.columns
vif
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled,y,test_size=.25,random_state=6)
```

```python
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc.score(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc.score(x_train,y_train)
```

```python
dtc.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(dtc,x_train,y_train,cv=1000)
```

```python
score.mean()
```

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
fin=pd.DataFrame({'current':y_test,
                 'predict':y_pred})
```
